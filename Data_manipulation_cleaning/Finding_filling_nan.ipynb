
# Data manipulation:

## Finding and filling NaN values

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

#### Finding NaN values

```python
blank = np.nan

serie = Series(['row1','row2', blank,'row4',
                'row5','row6',blank,'row8'])
serie
```

#### Identifies NaN values

```python
serie.isnull()
```

#### Filling nan values

```python
np.random.seed(25)
df = DataFrame(np.random.rand(36).reshape((6,6)))
df
```

##### df.loc[row:row, column]

```python
df.loc[1:3,5] = blank
```

```python
df.loc[1:4,0] = blank
df
```

```python
df_full = df.fillna(0)
df_full
```

#### Filling with different values

dicionary = {column: value, column: value, ...}

```python
dicio = {0: 0.1, 5: 1.25}
df_full = df.fillna(dicio)
df_full
```

```python
dicio = {0: np.mean(df[0]), 5: np.mean(df[5])}
df_full = df.fillna(dicio)
df_full
```
